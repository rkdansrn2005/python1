### 인공신경망

#### 1. 인공신경망의 개요

1. 개요
* 인공지능(AI)의 한 분야
* 사람의 두뇌의 생물학적 작동형태를 모방하여 컴퓨터로 하여금 지적 능력을 갖추게하는 방법론
* 공학부문에서 시작되어서 재무관리등에 도입되기 시작
* 인공신경망이 좀 더 확장된 것이 딥러닝(Deep Learning)이다.

2. 선형분류와 비선형 분류회귀분석은 선형적인 분류방법이지만 비선형적인 자료를 분석하기 위해서는 선형적인 방법으로는 분석하기 힘들기 때문에 비선형적인 즉, 인공신경망을 통한 분석방법이 필요하게 되었다.

3. 인공신경망의 구조 

> 3개의 층(Three Layer Network)
* 입력층(Input Layer) : 자료가 입력이 되면 
* 은닉충(Hidden Layer) : 은닉층에서 가중치등을 주어서 계산하고
* 출력충(Output Layer) : 출력층에서 그 결과를 받아서 출력
* 딥러닝(Deep Learning)은 은닉층의 갯수를 많이 늘려서 계산을 한다.
* H/W가격과 성능이 많이 개선이 되었기 때문에 가능해 졌다.



#### 2. 기본개념
1. 처리요소(PE)
* 각각의 인자(Neuron)을 의미
* 노드(Node) 또는 유니트(Unit)라고 한다.

2. 층(Layer)

* 보통 3개의 층(Three-Layer)
* 은닉층(Hidden) : 입력값과 출력값을 연결 시켜주는데 매개변수 값으로 채워지기 되는데 전체 신경망의 성공여부가 바로 이 은닉층에 달려 있다.

3. 가중지(Weight)

* i번째 PE와 j번째의 PE와의 연결 결합 정도 
* Wij : 층과 층사이의 연결이 i에서 j로 갈 때의 연결 강도 

#### 3. 다중퍼셉트론

1. 입력층과 출력층사이의 중간층 즉, 은닉층이 존재하는 신경망
1. 다중퍼셉트론에서의 가중치(Weight)
   * 지속적으로 전체 신경망이 만족할 만한 목표에 도달할 때까지 변함
     - 목표값이 도달할 때가 지속적으로 계산하기 때문에 계산량이 아주 많음
   * 계산된 출력값과 목표출력값(Target Output)을 비교하여 그 차이(오차함수)를 최소화 시킬 수 있도록 지속적으로 조정이 된다.
   * 학습방법 : 역전파 알고리즘(Backpropagation Algorithm)
     - 입력층의 각 유닛에 입력패턴 입력
     - 입력신호는 각 유닛에서 변환되어 중간층(은닉층)에 전달되고 최후에는 출력층에서 신호를 출력
     - 이 출력값과 기대값(목표값)을 비교하여 차이를 줄여나가는 방향으로 연결강도를 조절
     - 상위층에서 역전파하여 하위층에서는 다시 자기층의 연결강도(가중치)를 조정

#### 4.  인공신경망의 추론과정

1. 1단계 : 초기 연결가중치를 결정 : 연결가중치를 임의의 아주 작은 값(보통 -1~1)으로 초기화  
1. 2단계 : 전방향계산
   * 은닉층 및 출력층에서 입력값에 연결 가중치를 곱하여 각 처리요소(PE)들의 출력값을 계산
   * 전이함수를 사용하여 출력값을 결정
1. 3단계 : 역방향 계산
   * 츨력층의 출력값 목표출력값 사이의 오류치 계산
   * 출력층과 은닉층 사이의 연결 가중치를 수정
   * 은닉층과 입력층 사이의 연결 가중치를 수정
1. 4단계 : Epoch(2,3단계)의 반복
#### 5. 인공신경망의 학습절차

1. 1단계 : 자료수집 - 정확하고 충분한 자료의 수집이 필요(최소 1000건이상 필요)
1. 2단계 : 자료준비
   * 비계량화자료를 계량화
   * Outlier를 제거
   * 학습용자료와 검증용자료를 정의(training, test, validateion set)
   * 데이터전처리과정
1. 3단계 : 신경망의 구조와 학습모수의 결정
   * 은닉층의 수와 은닉노드의 수를 결정
   * 학습틀, 학습중지점등을 결정
1. 4단계 : 학습과 검증
   * 검증방법의 결정
   * 검증결과의 해석

##### 실습 1. 신경망튜닝

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

# 데이터셋 
X, y = datasets.make_moons(n_samples=100,noise=0.25,random_state=3)
X_train, X_test,y_train,y_test= train_test_split(X,y,stratify=y,random_state=42)

# 산점도
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
# 기계학습 : solver='lbfgs'은 가중치를 최적화 시키기 위한 알고리즘
mlp = MLPClassifier(solver='lbfgs', activation="tanh",random_state=0).fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=0.3)
plt.xlabel("특성0")
plt.ylabel("특성1")
plt.title("은닉층=1개, 은닉유닛=100개인 신경망 분류결과")
plt.show()
```

```python
# 한개의 은닉층 10개의 유닛에 
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
mlp = MLPClassifier(solver='lbfgs', activation="relu"
                    , hidden_layer_sizes=[10] ,random_state=0).fit(X_train, y_train)
# hidden_layer_sizes [은닉층(한층)의 크기,은닉층의 층의 갯수]
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=0.3)

plt.xlabel("특성0")
plt.ylabel("특성1")
plt.title("은닉층=1개, 은닉유닛=10개인 신경망 분류결과")
plt.show()

# 결과분석
# 은닉유닛이 10개이어서 결정경계가 100개보다 날카로워 졌다. 비선형 함수(활성화함수)
# 렐루 함수이다. 은닉층이 하나이기 때문에 결정경계를 만드느 함수는 직선 10개가 합쳐서 구성된다.
# 좀더 매끄러운 경계를 원한 다면 은닉유닛을 추가하거나 은닉층을 추가하거나 tanh함수를 
# 사용할 수 있다.
```

```python
# 두개의 은닉층 10개의 유닛에 
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
mlp = MLPClassifier(solver='lbfgs', activation="relu"
                    , hidden_layer_sizes=[100,100,10,10] ,random_state=0).fit(X_train, y_train)
# hidden_layer_sizes=[10,10,10,10] 4개의 은닉층 각 10개의 은닉유닛 보유
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=0.3)

plt.xlabel("특성0")
plt.ylabel("특성1")
plt.title("은닉층=2개, 각 은닉유닛=10개인 신경망 분류결과")
plt.show()
```

```python
# 은닉층과 은닉 유닛에 따른 결정경계
# 신경망의 복잡도를 제어하는 방법은 은닉층의 수, 은닉층의 유닛수, 규제(alpha)를 사용한다.
fig, axes=plt.subplots(2,4,figsize=(20,8))
for ax, n_hidden_nodes in zip(axes, [10,100]):
    for a, alpha in zip(ax, [0.0001,0.01,0.1,1]):
        mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,ax=a,s=5)

        mlp = MLPClassifier(solver='lbfgs', activation="relu"
                    , hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes],alpha=alpha ,random_state=0).fit(X_train, y_train)
        mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,ax=a,alpha=0.3)
        
        a.set_title("은닉층=[{},{}]\n알파값(규제)={:4f}".format(n_hidden_nodes,n_hidden_nodes,alpha))
        
plt.show()
```

```python
# 무작위로 서로 다른 초기값(데이터)을 부여하고 동일 매개변수로 학습한 결정경계
fig,axes = plt.subplots(2,4,figsize=(20,8))
for i, ax in enumerate(axes.ravel()):# axes.ravel()의 의미는 2x4의 그래프를 8개의 1차원 그래프로 변환
    mlp = MLPClassifier(solver='lbfgs', activation="relu"
                    , hidden_layer_sizes=[100, 100] ,random_state=i).fit(X_train, y_train)
    mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,ax=ax)
    mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,ax=ax,alpha=0.3)

plt.show()

    
    
```

#### 장단점과 매개변수

장점은 대량의 데이터에 내재된 정보를 찾아내고 매우 복잡한 모델을 만들 수 있다. 충분한 연산시간과 데이터를 주고 매개변수를 세심하게 조정하면 인공신경망은 (분류와 회귀에 모두) 종종 다른 머신러닝 알고리즘보다 뛰어난 성능을 나타낸다.

단점은 인공신경망은 (특히 크고 강력한 모델이거나 매개변수에 따라 )종종 학습시간이 오래걸린다. 또한 데이터 전처리에 주의해야 한다.

#### slover 매개변수
`solver`매개변수를 사용해 모델을 학습시키는 방법 또는 매개변수 학습에 사용하는 알고리즘을 사용자가 지정할 수가 있다. `solver`매개변수에는 쉬게 사용할 수 있는 옵션 두가지가 있는데 데 기본 값은 `adam`으로 대부분의 경우 잘 작동하지만 데이터의 스케일에 조금 민감하다. 그래서 데이터를 `평균0,분산1`로 조정하는 것이 중요하다.

다른 하나는 `lbfg`로 안정적이지만 규모가 큰 모델이나 대량의 데이터셋에서는 시간이 오래 걸린다. 또 다른 알고리즘은 많은 딥러닝 연구자가 사용하는 `sgd`가 있다. 이 옵션은 다른 여러 매개변수와 함께 튜닝하여 최선의 결과를 만들 수 있다.

#### 인공신경망 실습

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

# n_samples == 데이터개수, noise= 표준편차,r
X, y = datasets.make_moons(n_samples=1000,noise=0.25,random_state=3)

# 훈련용 vs 검증용 = 8 : 2
# stratify = y : 훈련용과 학습용 데이터에서 각각의 타겟값들이 동일한 비율로 나타내도록 함
# test_size : 검증용 자료가 20%인 의미 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2\
                                                   ,stratify=y, random_state=42)
# 기계학습
mlp =MLPClassifier(solver="lbfgs",random_state=0).fit(X_train,y_train)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
mglearn.plots.plot_2d_separator(mlp, X_train,fill=True, alpha=0.3)
plt.xlabel('특성0')
plt.ylabel('특성1')
plt.show()
```

```python
mlp =MLPClassifier(solver='adam',activation="relu"random_state=0).fit(X_train,y_train)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
mglearn.plots.plot_2d_separator(mlp, X_train,fill=True, alpha=0.3)
plt.xlabel('특성0')
plt.ylabel('특성1')
plt.show()
```

```python
# hidden layer를 2개, 각 은닉층마다 유닛수를 10개등등 
# 최적의 알고리즘 즉, solover옵션을 adam, sgd, lbfgs...변경해보기\n",
# 기개학습후 그래프로 시각화하기
X, y = datasets.make_moons(n_samples=1000,noise=0.25,random_state=3)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2\
                                                   ,stratify=y, random_state=42)

mlp =MLPClassifier(solver="sgd",activation="relu"
                   , hidden_layer_sizes=[10, 10],random_state=0).fit(X_train,y_train)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
mglearn.plots.plot_2d_separator(mlp, X_train,fill=True, alpha=0.3)
plt.xlabel('특성0')
plt.ylabel('특성1')
plt.show()
```

```python
# 훈련용과 검증용 = 75:25로 나누어서

X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,test_size=0.25,
                                                 stratify=cancer.target, random_state=42)

# 기계학습
mlp =MLPClassifier(solver="adam",activation="relu",hidden_layer_sizes=[10,10],random_state=0).fit(X_train,y_train)
# 결과출력

print("훈련용 세트 정확도 : {:.3f}".format(mlp.score(X_train,y_train)))
print("검증용 세트 정확도 : {:.3f}".format(mlp.score(X_test,y_test)))

# 결과분석
# 훈련용은 92%의 정확도와 검증ㅇ용은 93%의 정확도를 가진다.
# 해당 데이터셋의 전처리를 전이기 때문에 전처리후의 결과와 비교할 필요가 있다.
```

```python
# 데이터의 전처리 : -1.0~1.0사이의 값으로 전처리
# StandardScaler함수로 사용하지 않고 실습에서는 수작업으로 스케일 작업을 진행

# 전처리방법
# 1. 훈련세트의 각 특성으 평균을 계산
mean_ontrain = X_train.mean(axis=0)

# 2. 훈련세트의 각 특성의 표준편차계산
std_ontrain = X_train.std(axis=0)
# print(mean_ontrain)
# print(std_ontrain)

# 3. 데이터에서 평균을 빼고 표준편차로 나누면 
#    평균0, 표준편차가 1인 데이터로 변환이 된다.
X_train_scaled = (X_train-mean_ontrain)/std_ontrain
# print(X_train_scaled)

# 4. 훈련용 데이터셋에도 동일 계산을 적용
mean_ontest = X_test.mean(axis=0)
std_ontest = X_test.std(axis=0)
X_test_scaled = (X_test - mean_ontest)/std_ontest
# X_test_scaled = (X_test - mean_ontrain)/std_ontest


# 5. 기계학습
mlp =MLPClassifier(random_state=42).fit(X_train_scaled,y_train)

# 결과 출력
print("훈련용 세트 정확도 : {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("검증용 세트 정확도 : {:.3f}".format(mlp.score(X_test_scaled,y_test)))

X_test_scaled = (X_test - mean_ontrain)/std_ontrain
print("훈련용 세트 정확도 : {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("검증용 세트 정확도 : {:.3f}".format(mlp.score(X_test_scaled,y_test)))
# 7. 결과 분석
# 스케일 조정 즉 데이터의 표준화에 따라 성능이 많이 향상된 결과를 볼 수 있다.

```

```python
# 최적화 알고리즘을 변경(adam -> lbfgs)
mlp =MLPClassifier(solver="lbfgs",random_state=42).fit(X_train_scaled,y_train)

print("훈련용 세트 정확도 : {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("검증용 세트 정확도 : {:.3f}".format(mlp.score(X_test_scaled,y_test)))
```

```python
# 최적화 알고리즘을 변경(adam -> lbfgs)
mlp =MLPClassifier(solver="adam",max_iter=1000,random_state=42).fit(X_train_scaled,y_train)
# max_iter=200 200번을 반복
print("훈련용 세트 정확도 : {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("검증용 세트 정확도 : {:.3f}".format(mlp.score(X_test_scaled,y_test)))
```

```python
# 규제(alpha=0.0001)를 1로 증가(가중치를 더 강하게 규제)
mlp =MLPClassifier(solver="adam",max_iter=1000,alpha=1,random_state=42).fit(X_train_scaled,y_train)

print("훈련용 세트 정확도 : {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("검증용 세트 정확도 : {:.3f}".format(mlp.score(X_test_scaled,y_test)))
```

#### 실습2: 붓꽃/Perceptron
iris데이터셋을 Perceptron 인공신경망 머신러닝을 적용해 보기 

* 퍼셉트론은 모든 학습데이터를 정확히 분류시킬 때까지 학습이 진행이 되기 때문에 학습데이터가 선형적으로 분리될 수 있을 때 적합한 알고리즘이다.

```python
iris = datasets.load_iris()
idx = np.in1d(iris.target,[0,2]) # in1d 첫번째 배열 x가 두번째 배열 y의 원소를 포함하고 있는지 여부의 불리언 배열을 반환

# idx
X = iris.data[idx,:2] # 0,1열 꽃받침 부위
# X
# print(iris.DESCR)
y = iris.target[idx]

plt.figure(figsize=(12,8),dpi=60)
plt.scatter(X[:,0], X[:,1], c=y, s=100, edgecolors='k')
plt.show()
```

```python
from sklearn.linear_model import Perceptron
import seaborn as sns
import matplotlib as mpl

def plot_perceptron(n):
    plt.figure(figsize=(12,8),dpi=60)
    model = Perceptron(max_iter=n,random_state=1).fit(X,y)
    XX_min = X[:,0].min() - 1;
    XX_max = X[:,0].max() + 1;
    YY_min = X[:,1].min() - 1;
    YY_max = X[:,1].max() + 1;
    XX, YY = np.meshgrid(np.linspace(XX_min, XX_max,1000),np.linspace(YY_min, YY_max,1000))
#     meshgrid의 목적은 x 값의 배열과 y 값의 배열로 직사각형 격자를 만드는 것입니다
#     print(XX)
#     print(YY)
    ZZ = model.predict(np.c_[XX.ravel(),YY.ravel()]).reshape(XX.shape)
#     print(ZZ)
    cmap = mpl.colors.ListedColormap(sns.color_palette("Set2"))
    plt.contourf(XX,YY,ZZ,cmap=cmap)
    plt.scatter(X[:,0], X[:,1], s=50, linewidths=2, edgecolors='k', c=y, cmap=cmap)
    plt.xlim(XX_min,XX_max)
    plt.ylim(YY_min,YY_max)
    plt.grid(False)
    plt.show()
```

```python
plot_perceptron(1)
```

```python
from ipywidgets import widgets
widgets.interact(plot_perceptron,n=widgets.IntSlider(min=1,max=500,step=1,value=1))
```
