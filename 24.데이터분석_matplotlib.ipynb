#### Matplotlip에서 그래프 포맷

색상을 지정하는 방법은 색이름 또는 약자를 사용하거나 RGB코드(#)를 사용한다. 자주 사용되는 경우 한 문자의 약자로 사용한다.

* 참고 :http://matplotlib.org/examples/color/named_colors.html

#### 색상별 약자
> * blue -b
* green = g
* red = r
* cyan = c
* magenta = m
* yellow = y
* black = k
* white = w


```python
from __future__ import division

import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols + 1

fig, ax = plt.subplots(figsize=(8, 5))

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
    col = i % ncols
    row = i // ncols
    y = Y - (row * h) - h

    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)

    ax.text(xi_text, y, name, fontsize=(h * 0.8),
            horizontalalignment='left',
            verticalalignment='center')

    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(h * 0.6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```

##### Figure 객체
모든 그래프에는 `Figure객체`, 정식으로는 `Matplotlib.figure.Figure`클래스 객체가 포함되어 있다.
내부 그래프가 아닌 경우에는 하나의 Figure는 하나의 윈도우를 가진다.
주피터 노트북에서는 윈도우객체가 생성되지 않지만 파이썬을 독립적으로 실행할 경우에는 하나의
Figure당 하나의 아이디와 윈도우를 별도를 가진다.

원래는 Figure를 생성하려면 figure라는 명령을 사용해서 Figure객체를 얻어야 하지만 일반적으로
`plot명령등을 실행하면 자동으로 Figure를 생성`해 주기 때문에 figure명령을 잘 사용하지는 않는
다. figure명령을 명시적으로 사용하는 경우에 `여러개의 윈도우를 동시에 띄우거`나 주피터 노트북
에서 `그림의 크기를 설정`할 경우 사용한다.

```python
# 그림 크기 설정하기
np.random.seed(0)
f1 = plt.figure(figsize=(10,1))

plt.title("그림크기설정하기")
plt.plot(np.random.randn(100))
plt.show()
```

##### Axes객체와 subplot 명령

하나의 윈두우(Figure)안에 여러개의 그래프를 배열형태로 보여야 할 경우에 Figure안에 있는각각의 그래프는 `Axes라고 불리는 객체`이다.
Figure안에 Axes를 생성하려면 `subplot명령`을 사용해서 명시적으로 Axes객체를 얻어야 한다. 그러나 plot명령을 바로 사용해도 자동으로 Axes객체를 생성해 준다.

subplot명령은 그리드(grid)형태의 Axes객체를 생성하는데 `Fiture가 행렬(Matrix)`이고 `Axes가 행렬의 원소`라고 생각하면 된다. 주의할 점은 첫번째 `Axes객체를 가리키는 숫자는 0이 아니라
1인 것에 주의` 해야 한다.
<img src='./images/24.데이터분석_matplotlib_01.png' width=300 height=200/>

```python
# 1. subplot(x,y,z)
# 두개의 그래프를 그릴 경우 행이 2인 열1, 즉 2행 1열인 행렬이다. subplot 명령은
# 3개의 인수값을 가진다. 처음 인수가 행, 두번째 인수가 열, 세번째 인수는 
# Axes객체를 의미한다. 예를 들어서 subplot(2,2,1), subplot(222)형식으로 정의한다.
# plt.subplot?

x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

# Axes 객체 생성
ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')

plt.tight_layout() # 그래프간의 간격을 자동으로 맞춤
plt.show()
```

```python
# 에노데이션
# 그래프에 화살표를 그린 후 그 화살표에 문자열을 출력하는 기능이다.
# 예를 "이 값이 최소값"이런 식으로 화살표를 그려서 표현하고자 할 때 사용하는 기능
# plt.annotate('문자열',xy,xytext,arrowprops)식으로 사용한다.
# 문자열은 어노테이션 나타날 문자열이고, xy,는 화살표가 가르키는 점의 위치, xytext는 
# 문자열이 출력될 위치, arrowprops는 화살표의 속성으로 컬러등을 정의한다.
# plt.annotate?
x = np.arange(1,10)
y = x*5
plt.plot(x,y)
plt.annotate('이 곳이 최소값',xy=(2,10),xytext=(5,20),arrowprops={'color':'red'})
plt.show()
```

```python
# Gradient descent
# 경사하강(기울기하강)이란? 특정 위치에대한 스칼라함수가 존재할 때 그 위치의 주변에서
# 가장 함수의 값을 커지는 방향(최적의 값)을 가르키는 벡터이다.
# 벡터는 방향과 크기(길이)가 있는 값이라면 스칼라는 값만 있는 자료이다. 여기서 이야기하는
# 벡터는 화살표를 의미한다. 또한 gradient벡터의 크기는 기울기 값이다.
# f(x) = x2 이라는 함수에서의 gradient를 그리기
x = np.linspace(-2,3,100)
y = x**2

fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y,color='black',linestyle="-",lw=3)
ax.scatter(2,4,color='y',s=500,alpha=0.5)
ax.arrow(2,4,0.4,1.6,lw=3,head_width=0.2,head_length=0.1,fc='red',ec='red')
ax.arrow(2,4,-0.4,-1.6,lw=3,head_width=0.2,head_length=0.1,fc='blue',ec='blue')
ax.scatter(2-0.1*4,(2-0.1*4)*2,color='y',s=500,alpha=0.5)

plt.show()
```

##### 연습문제  : 지역별 인구 밀도(2017년)
<img src='./images/24.데이터분석_matplotlib_02.png' width=300 height=200/>

```python
# 도시
cities=['서울','인천','대전','대구','울산','부산','광주']
# 2017년 통계청자료 인구밀도(명/Km^2)
pop = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

# 위도경도
size = []
for i in pop:
    size.append(i/10)
colors = ['r','g','b','c','m','k','y']
lat_lon={37.56667:126.97806, 37.45639:126.70528, 36.35111:127.38500, 35.87222:128.60250,
        35.53889:129.31667,35.17944:129.07556,35.15972:126.85306}
lat = list(lat_lon.keys())
lon = list(lat_lon.values())
for i in range(len(cities)):
    plt.text(lon[i],lat[i],cities[i])
plt.scatter(lon,lat,c=colors,s=size)
plt.xlabel('경도')
plt.ylabel('위도')
plt.title("인구밀도")
plt.grid()
plt.savefig('파일이름.png')
plt.show()
```

```python
df_rain = pd.read_csv('../data/db/sea_rain.csv', index_col='연도', encoding='utf-8')
df_rain1 = pd.DataFrame()
df_rain1 = df_rain1.append(df_rain)
year=[1996,1997,1998,1999,2000]

east_sea = list(df_rain1.iloc[:,0])
south_sea = list(df_rain1.iloc[:,1])
west_sea = list(df_rain1.iloc[:,2])
all_sea = list(df_rain1.iloc[:,3])
print(west_sea)
```

```python
plt.plot(year, east_sea,'r-^', label='동해')
plt.plot(year, south_sea, 'g-', label="남해")
plt.plot(year, west_sea, 'c-.d', label="서해")
plt.plot(year, all_sea, 'm:*', label="전체")
plt.grid()
plt.legend(loc=0)
plt.xlabel('년도')
plt.ylabel("강수량")
plt.title("년간 해상 강우량")
plt.show()
```
